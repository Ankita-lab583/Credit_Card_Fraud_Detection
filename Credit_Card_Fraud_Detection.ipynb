Import Dependecies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

loading dataset to a panda dataframe

```python
crd_data = pd.read_csv('/content/creditcard.csv')
```

There are discripencies in this dataset.
using  0--> Normal Transaction
and    1--> Fraudulent transaction

#separating the data for analysis

```python
legit = crd_data[crd_data.Class == 0]
fraud = crd_data[crd_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#getting statistical measures of data
legit.Amount.describe()
```

```python
#getting statistical measures of data
fraud.Amount.describe()
```

```python
#comparing the values for bith transactions
crd_data.groupby('Class').mean()
```

```python
#building a sample dataset containing similar distribution of normal transactions and fraudulent transaction
#Number of fraudulet transactions--> 392
legit_sample = legit.sample(n = 392)
```

```python
#Concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```


splitting the dataset into Features and Targets


```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

#splitting of dataset into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify= Y, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Model training by using Logistic Regression

```python
model = LogisticRegression()
```

```python
#traing the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

#Model Evalution: Accuracy Score

```python
#accuracy of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy of training data : ", training_data_accuracy )
```

```python
#accuracy of testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of testing data : ", testing_data_accuracy )
```
